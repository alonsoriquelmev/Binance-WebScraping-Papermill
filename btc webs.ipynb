```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import requests as req
from time import sleep
from selenium.webdriver.common.keys import Keys
import pandas as pd
import datetime as dt
import openpyxl
import os
```

```python
driver = webdriver.Chrome('chromedriver.exe')
url = 'https://www.binance.com/es/markets/spot'
driver.get(url)
web_element = driver.find_elements_by_xpath("//*[@id='market_trade_list_item']")
currencies = []
prices = []
clean_prices = []
for element in web_element:
        currency = element.find_element_by_class_name('css-17wnpgm').text
        price = element.find_element_by_class_name('css-1vzayxe').text
        currencies.append(currency)
        prices.append(price)
dict_replace = {'/ â‚¬' : '', '.' : '', ',' : '.'}
for i in range(len(prices)):
    price_aux = prices[i]
    for k,v in dict_replace.items():
            price_aux = price_aux.replace(k,v)
    
    clean_prices.append(price_aux)
sleep(3)
driver.close()

data_dict = {'Currency' : currencies, 'Price' : clean_prices, 'TimeStamp' : dt.datetime.today()}
df_cons = pd.DataFrame(data_dict)
df_cons['Price'] = df_cons['Price'].astype(float)
df_cons = df_cons.head(15)

path = 'Cryptocurrencies.xlsx'

try:
    with pd.ExcelWriter(path,engine = 'openpyxl',mode = 'a',if_sheet_exists = 'overlay') as writer:
        df_cons.to_excel(writer, sheet_name = 'Binance', startrow = writer.sheets['Binance'].max_row, index = False, header = None)
except:
        df_cons.to_excel(path,sheet_name = 'Binance',index = False, header = None)
```
